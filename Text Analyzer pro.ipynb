```python
name = (input("Enter the sentence"))

count = 0

for char in name.lower():
    if char in "a,e,i,o,u":
        count += 1

print("Vowels: ",count)
```

```python
name = "i love data science"
word = input("Enter the word to find: ")

if word in name:
    print("The Founded word index is: ", name.find(word))
else:   
    print("The word is not found in the Sentence.")
 
```
